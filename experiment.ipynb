````
import os
import shutil
import time

# Create an original evidence file
with open("evidence_original.txt", "w") as f:
    f.write("Case File #001 - Suspect device log")

# Get original file metadata
orig_stat = os.stat("evidence_original.txt")

print("Original file")
print("Created (ctime):", time.ctime(orig_stat.st_ctime))
print("Modified (mtime):", time.ctime(orig_stat.st_mtime))

# Simulate copying the file
time.sleep(1)
shutil.copy2("evidence_original.txt", "evidence_copy.txt")

# Get copied file metadata
copy_stat = os.stat("evidence_copy.txt")

print("\nCopied file")
print("Created (ctime):", time.ctime(copy_stat.st_ctime))
print("Modified (mtime):", time.ctime(copy_stat.st_mtime))

print("\nLocard's Exchange Principle:")
print("The copy operation created a new creation timestamp")
print("on the destination file. Every interaction leaves a trace.")
````

````
import hashlib
import datetime


def sha256_of(filename):
    with open(filename, "rb") as f:
        return hashlib.sha256(f.read()).hexdigest()


# Create evidence
with open("evidence.txt", "w") as f:
    f.write("Suspect chat log: meeting at 10pm, bring the drive.")

# Initial hash
seizure_hash = sha256_of("evidence.txt")

custody_log = []

custody_log.append(
    f"{datetime.datetime.now()} - SEIZED by Officer A - Hash = {seizure_hash}"
)

# Verify integrity during handoff
handoff_hash = sha256_of("evidence.txt")

if handoff_hash == seizure_hash:
    custody_log.append(
        f"{datetime.datetime.now()} - RECEIVED by Analyst B - Integrity VERIFIED"
    )
else:
    custody_log.append(
        f"{datetime.datetime.now()} - RECEIVED by Analyst B - Integrity FAILED"
    )

print("=== Chain of Custody Log ===\n")

for entry in custody_log:
    print(entry)
````

````
import re


def check_email(sender, subject, body):
    flags = []

    if re.search(r"(urgent|verify your account|suspended|click here)", body, re.IGNORECASE):
        flags.append("Urgency or pressure language detected")

    if re.search(r"\d{1,3}(\.\d{1,3}){3}", body):
        flags.append("Raw IP address found")

    domain = sender.split("@")[-1]

    if any(word in domain.lower() for word in ["secure", "verify", "update"]) and "@gmail" not in sender:
        flags.append("Suspicious sender domain")

    return flags


emails = [
    (
        "support@paypal.com",
        "Your monthly statement",
        "Please find your statement attached."
    ),
    (
        "alert@paypal-secure-verify.com",
        "URGENT: Verify your account",
        "Click here: http://192.168.10.5/login"
    )
]

for sender, subject, body in emails:

    issues = check_email(sender, subject, body)

    verdict = "PHISHING SUSPECTED" if issues else "Looks Legitimate"

    print("\nFrom:", sender)
    print("Subject:", subject)
    print("Verdict:", verdict)

    for issue in issues:
        print("-", issue)
````

````
import hashlib

def sha256_of(filename):
    with open(filename, "rb") as f:
        return hashlib.sha256(f.read()).hexdigest()


# Create a sample disk image
with open("original_disk.img", "wb") as f:
    f.write(b"HEADER" + bytes(range(256)) * 4 + b"FOOTER")

# Create a forensic (bit-stream) copy
with open("original_disk.img", "rb") as src, open("forensic_copy.img", "wb") as dst:
    dst.write(src.read())

# Verify integrity
original_hash = sha256_of("original_disk.img")
copy_hash = sha256_of("forensic_copy.img")

print("Original Image Hash :", original_hash)
print("Forensic Copy Hash  :", copy_hash)

if original_hash == copy_hash:
    print("\nVERIFIED - Exact Bit-Stream Copy")
else:
    print("\nMISMATCH - Copy Corrupted")
````

Experment number 7

````
SIGNATURES = {
    b"\xFF\xD8\xFF": "JPEG Image",
    b"\x89PNG": "PNG Image",
    b"%PDF": "PDF Document",
    b"PK\x03\x04": "ZIP Archive (DOCX/XLSX)"
}


def identify_file(path):
    with open(path, "rb") as f:
        header = f.read(8)

    for signature, filetype in SIGNATURES.items():
        if header.startswith(signature):
            return filetype

    return "Unknown File Type"


# Create sample files
with open("photo.txt", "wb") as f:
    f.write(b"\xFF\xD8\xFF\xE0" + b"\x00" * 20)

with open("document.dat", "wb") as f:
    f.write(b"%PDF-1.4" + b"\x00" * 20)

# Detect actual file types
files = ["photo.txt", "document.dat"]

for file in files:
    print(f"{file:15} -> {identify_file(file)}")
````

````
import os

# Create a sample file
with open("secret_note.txt", "w") as f:
    f.write("MEETING_POINT: warehouse 7, 11pm")

# Read original bytes
with open("secret_note.txt", "rb") as f:
    original_bytes = f.read()

# Delete the file
os.remove("secret_note.txt")

print("File deleted.")
print("File exists:", os.path.exists("secret_note.txt"))

# Simulate disk buffer
disk_buffer = original_bytes + b"\x00" * 50

marker = b"MEETING_POINT"

index = disk_buffer.find(marker)

if index != -1:
    recovered = disk_buffer[index:].split(b"\x00")[0]

    with open("recovered_note.txt", "wb") as f:
        f.write(recovered)

    print("\nRecovered Data:")
    print(recovered.decode())
else:
    print("No recoverable data found.")
````
